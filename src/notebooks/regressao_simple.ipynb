### Criando regressão linear

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("../dataset/regressao_simples.csv")
df.head()
```

```python
df.info()
```

```python
# Definindo Variáveis
x = df["Idade"]
y = df["Salario"]
print(x[0], y[0])
```

```python
x = df["Idade"].values.reshape(-1, 1) # Necessário a colunas ser 2D.
y = df["Salario"].values
# criando e treinando o modelo
model = LinearRegression()
model.fit(x, y)
# para visualizar o R²
model.score(x,y)
print(f"Seu score é: {model.score(x,y)}")
```

```python
previsao_idade = [[44]]
salario_previsto = model.predict(previsao_idade)
print(f"Salário previsto para idade {previsao_idade[0][0]}: R$ {salario_previsto[0]:.2f}")
```
